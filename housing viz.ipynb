```python
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import unicodedata
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By

driver = webdriver.Firefox()

apartments = []

for i in range(1,120):

    website = 'http://streeteasy.com/for-rent/nyc/' + str(i)
    driver.get(website)
    driver.implicitly_wait(5)

    rows = driver.find_elements_by_class_name('details-title')

    for row in rows:
       name = row.find_element_by_tag_name('a')
       apartments.append(name.text)

print apartments

#http://www.softqe.com/selenium-how-to-wait-for-page-load/
#https://stackoverflow.com/questions/30114976/the-default-value-of-timeouts-on-selenium-webdriver
```

```python
## clean the data from street easy by removing the apartment number after the building location
## then before printing return apartment if any match without apartment number

cleaned_apartments = []

for i in range(len(apartments)):
    str = apartments[i] 
    index = str.find('#')
    cleaned_apartments.append(str[:index])

cleaned_apartments = [x.encode('UTF8') for x in cleaned_apartments]
cleaned_apartments = [x[:-1] for x in cleaned_apartments]

print cleaned_apartments
```
